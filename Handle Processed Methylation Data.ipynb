## Methods start here

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./trans_processed_PPMI_data.csv')
data
```

```python
data.rename(columns={'Unnamed: 0':'Sentrix_position'}, inplace=True)
data.set_index('Sentrix_position', inplace=True)
ppmi = data.transpose()
ppmi
```

## Run Classifier on original unreduced data 

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
label = encoder.fit_transform(ppmi['Category'])
label
```

```python
tr = ppmi.drop(['Category'], axis=1)
X = tr.values
y = label
print(X.shape)
print(y.shape)
```

```python
#Stratified sampling
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
split.get_n_splits(X, y)

for train_index, test_index in split.split(X, y):
    print("TRAIN:", len(train_index), "TEST:", len(test_index))
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
### Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled
X_test_scaled = scaler.fit_transform(X_test)
```

### Classifiers

#### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
#######Train model using X_train_scaled directly without cross-validation
log_reg = LogisticRegression(max_iter=300)
log_reg.fit(X_train_scaled, y_train)
y_pred = log_reg.predict(X_test_scaled)
accuracy_score(y_test, y_pred)
```

#### Notes:
- cross_val_score: used as an analysis tool to evaluate the results obtained by training strategy used before (i.e. the model may be applied entirely)
- cross_val_predict: apply cross-validation to training and get predictions and use it for analysis

#### 2. SVM

```python
from sklearn.linear_model import SGDClassifier

#Here, no regularization applied when penalty set to None
sgd = SGDClassifier(tol=None, random_state=42,
                   eta0=1, penalty=None)
sgd.fit(X_train_scaled, y_train)
y_pred = sgd.predict(X_test_scaled)
accuracy_score(y_test, y_pred)
```

### Apply PCA to dataset and train classifiers on the principle components

```python
#######PCA on PPMI#########
import sklearn.decomposition

pca = sklearn.decomposition.PCA(n_components=80)
pca_train = pca.fit_transform(X_train_scaled)
pca_test = pca.fit_transform(X_test_scaled)
print(pca_train.shape)
print(pca_test.shape)
```

```python
#Run classifier on PCA components
#####Logistic Regressor########
log_reg_pca = LogisticRegression(max_iter=300)
log_reg_pca.fit(pca_train, y_train)
y_pred_pca = log_reg_pca.predict(pca_test)
accuracy_score(y_test, y_pred_pca)
```

```python
#######SVM######
sgd = SGDClassifier(tol=None, random_state=42,
                   eta0=1, penalty=None)
sgd.fit(pca_train, y_train)
y_pred = sgd.predict(pca_test)
accuracy_score(y_test, y_pred)
```

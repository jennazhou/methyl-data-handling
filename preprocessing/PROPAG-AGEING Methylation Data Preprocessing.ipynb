# Pre-processing PROPAG-AGEING dataset 

```python
import pandas as pd
import numpy as np
```

```python
propag_ageing = pd.read_csv('../datasets/PROPAG-AGEING/methylation_propagageing.csv', index_col='ID_REF')
# list(propag_ageing.columns)
propag_ageing = propag_ageing.drop(['Unnamed: 0'], axis=1)
```

```python
data = pd.read_csv('./trans_processed_PPMI_data.csv')
data
```

```python
genes = data[['Unnamed: 0']].values.flatten()
final_genes = np.delete(genes, -1)
final_genes.size
```

```python
propag_ageing
```

```python
ppg = propag_ageing.loc[final_genes].transpose()
ppg
```

```python
df = pd.read_csv('../datasets/PROPAG-AGEING/propag_phenotypes.csv')
df.set_index('Colname', inplace=True)
df
```

```python
df.groupby('Group').count()
```

```python
group = df[['Group']].values.flatten()
group
ppg['Category'] = group
ppg.dropna(subset=['Category'], inplace=True) 
ppg
```

```python
ls = ["CTR_GENERAL", "CTR_GENERAL_KI", "PD_D", "PD_TWIN_INCIDENT", "PD_TWIN_PREVALENT"]
ppg_final = ppg.loc[ppg["Category"].isin(ls)]
ppg_final.replace(['CTR_GENERAL', "CTR_GENERAL_KI"], "HC", inplace=True)
ppg_final.replace(["PD_D", "PD_TWIN_INCIDENT", "PD_TWIN_PREVALENT"], "PD", inplace=True)
ppg_final
```

```python
final_ppg_data_trans = ppg_final.transpose()
final_ppg_data_trans
```

```python
###save flipped data to .csv

final_ppg_data_trans.to_csv(r'./trans_processed_PPG_data.csv', header=True)
```

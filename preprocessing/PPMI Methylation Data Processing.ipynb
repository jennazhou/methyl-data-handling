## Filter out sex-related genes and impeding genes

```python
ppmi = pd.read_csv('../datasets/PPMI/beta_post_Funnorm_PPMI_EPICn524final030618.csv')
# ppmi
```

```python
propag_ageing = pd.read_csv('../datasets/PROPAG-AGEING/methylation_propagageing.csv', index_col='ID_REF')
# list(propag_ageing.columns)
propag_ageing = propag_ageing.drop(['Unnamed: 0'], axis=1)
# propag_ageing
```

```python
result = pd.concat([ppmi, propag_ageing], axis=1, join='inner', ignore_index=False)
result = result.rename_axis("Feature")
# result = result.rename_axis("Sample ID", axis="columns")
# result = result.transpose()
# result

#Table of the common features used in both datasets
```

```python
common_genes = result.index.values
common_genes
```

```python
del_features = pd.read_csv('../datasets/annot_info.csv')
del_features
```

```python
####del_features contain sex genes

del_features['ID_REF'][0][:2]
# print(del_features['CHR'].unique())
# del_features[(del_features['ID_REF'].str[:2]) == 'cg']
X_chr = del_features[del_features['CHR'] == 'X']
print("Size of sex genes with X_chr:" + str(X_chr.size))
Y_chr = del_features[del_features['CHR'] == 'Y']
print("Size of sex genes with Y_chr:" + str(Y_chr.size))
sex_genes = pd.concat([X_chr, Y_chr])
sex_genes = sex_genes['ID_REF'].values
sex_genes
```

```python
impeding_genes = pd.read_table('../datasets/EPIC_probes_to_be_filtered.txt', delim_whitespace=True)
impeding_genes = impeding_genes['EPIC_probes_to_be_filtered'].values
print(impeding_genes)
print('Number of common genes:' + str(common_genes.size))
print('Number of impeding genes:' + str(impeding_genes.size))
print('Number of sex genes:' + str(sex_genes.size))
```

```python
set1 = set(common_genes.flatten())
set2 = set(sex_genes.flatten())
set3 = set(impeding_genes.flatten())


# df1 = pd.DataFrame(data=common_genes[:], columns=['Features'])
# print(df1.size)
# df2 = pd.DataFrame(data=impeding_genes[:], columns=['Features'])
# # print(df2)
# df3 = pd.DataFrame(data=sex_genes[:], columns=['Features'])
# print(df3.shape)
# #remove sex genes from the dataset
# sexgenes_toberemoved = pd.merge(df1, df3, how='inner', on=['Features'])['Features'].values
# print("Number of sex genes to be removed from PPMI dataset: " + str(sexgenes_toberemoved.size))


#firstly, remove sex related genes from original dataset
sgremoved_set = set1.difference(set2)
sexgenes_removed = np.array(list(sgremoved_set))
print(sexgenes_removed.size)
sexgenes_removed

#secondly, remove impeding genes from the remaining
igremoved_set = sgremoved_set.difference(set3)
final_genes = np.array(list(igremoved_set))
final_genes.size
# igenes_removed
```

```python
selected_ppmi = ppmi.loc[final_genes]
selected_ppmi = selected_ppmi.transpose()
selected_ppmi
```

## Re-grouping for PPMI dataset

```python
pd.read_csv('../datasets/PPMI/EstimateCellCounts_PPMI_EPICn524final030618.csv')
```

```python
patinfo = pd.read_csv('../datasets/PPMI/Patient_Status.csv', index_col='PATNO')
# patinfo.sort_values(by='PATNO')
patinfo.index.astype('int64')
patinfo
```

```python
patno = pd.read_csv('../datasets/PPMI/PPMI_Meth_n524_for_LONI030718.txt',  sep="\t", header=None)
patno.columns = ["PATNO", "Sentrix ID", "Sentrix Position"]
patno.drop([0], axis=0, inplace=True)
patno.set_index(pd.to_numeric(patno["PATNO"]), inplace=True)
patno.drop(columns=["PATNO"], inplace=True)
patno
```

```python
#Get the info for required patients. Only selected enroll_category
result = pd.concat([patinfo, patno], axis=1, join='inner', ignore_index=False)[['ENROLL_CAT']]
result.reset_index("PATNO",inplace=True)
print("Result table with patient number and enroll_category: \n")
print(result)
ppmi_classes_group = result.groupby('ENROLL_CAT').count()
# ppmi_classes_group.index
ppmi_classes_group.rename(columns={"PATNO":"Number of patients"}, inplace=True)
ppmi_classes_group
```

```python
#result contains 524 rows, including SWEDD, with columns: PATNO, ENROLL_CAT
ppmi_labels = result["ENROLL_CAT"].values
patno["ENROLL_CAT"] = ppmi_labels
display(patno)

#ppmi_patno_woswedd only contains 470 rows
ppmi_patno_woswedd = patno[patno["ENROLL_CAT"]!="SWEDD"]
print("Number of patients without an enrollment category: " + str(ppmi_patno_woswedd[ppmi_patno_woswedd["ENROLL_CAT"].isnull()].shape[0]))

#drop null categories
ppmi_patno_woswedd.dropna(inplace=True)
display(ppmi_patno_woswedd)

#sentrix_position combines two columns to form the index of the big ppmi table
sentrix_position = ppmi_patno_woswedd["Sentrix ID"]+"_"+ppmi_patno_woswedd["Sentrix Position"]
ppmi_patno_woswedd["Sentrix_position"] = sentrix_position
display(ppmi_patno_woswedd)

#Now, we want a new dataframe that contains only Sentrix_position as index and ENROLL_CAT as labels which are to be matched with original ppmi table
ppmi_classes = pd.DataFrame(index=ppmi_patno_woswedd["Sentrix_position"], data={"Category":ppmi_patno_woswedd["ENROLL_CAT"].values})

```

```python
#samples with only HC and PD groups
selected_ppmi_hc_pd = selected_ppmi.loc[ppmi_classes.index]
selected_ppmi_hc_pd["Category"] = ppmi_classes["Category"]
selected_ppmi_hc_pd
```

## Select on Caucasians from the dataset

```python
demo = pd.read_csv('../datasets/PPMI/Screening_Demographics.csv')
demo
```

```python
#Get the match between PATNO and Sentrix Position
patno_num_sentrix = ppmi_patno_woswedd[['Sentrix_position']]
display(patno_num_sentrix)

#Get the match between races and PATNO
# demo.info()
races = demo[['PATNO','RAWHITE']].loc[demo['PATNO'].isin(patno_num_sentrix.index)].sort_values(by=['PATNO'])
# demo[['PATNO','RAWHITE']]
display(races)
numof_whites = races.groupby('RAWHITE').count()
numof_whites
```

```python
final_patno = races[races['RAWHITE'] == 1.0][['PATNO']].values.flatten()
final_sentrix = patno_num_sentrix.loc[final_patno][['Sentrix_position']].values.flatten()
final_ppmi_data = selected_ppmi_hc_pd.loc[final_sentrix]
final_ppmi_data
```

```python
####flip the dataframe and save for faster read
final_ppmi_data_trans = final_ppmi_data.transpose()
final_ppmi_data_trans
```

```python
###save flipped df to .csv
final_ppmi_data_trans.to_csv(r'./trans_processed_PPMI_data.csv', header=True)
```
